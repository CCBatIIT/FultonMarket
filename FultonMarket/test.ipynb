```python
class Parent():

    def __init__(self, a: str):
        self.a = a

    def run(self):
        self._print_statement()

        print('Completed')

    def _print_statement(self):
        print('Hello' + self.a)
```

```python
class Child(Parent):

    def __init__(self, a: str):
        super().__init__(a)

    def _print_statement(self):
        print('Hello' + self.a +'This shit worked')
```

```python
parent = Parent('test')
parent.run()
```

```python
child = Child('test2')
child.run()
```
