```python
pdbs = sorted(glob('CB2_centroids/inputs/*.pdb'), key = lambda x: int(x.split('/')[-1].split("_")[-1].split(".")[0]))
```

```python
sequences = {}

for i, pdb in enumerate(pdbs):
    traj = md.load(pdb)
    print(pdb)
    protein_inds = traj.top.select('protein')
    prot_top = traj.top.subset(protein_inds)
    resnames = [residue.name for residue in prot_top.residues]
    sequences[i+1] = resnames
```

```python
mapp = {'water': "blue", 'resname POP': "orange", 'protein': "green", "resname NA or resname CL": "red"}
```

```python
vals = []
for i, pdb in enumerate(pdbs):
    traj = md.load(pdb)
    traj_vals = []
    traj_vals.append(i+1)
    for selection in mapp.keys():
        traj_vals.append(traj.top.select(selection).shape[0])
    print(traj_vals)
    vals.append(traj_vals)
vals = np.array(vals)
```

```python
centroids = vals[:, 0]
n_waters = vals[:, 1]
n_membrane = vals[:, 2]
n_protein = vals[:, 3]
n_ion = vals[:, 4]
```

```python
plt.clf()
fig, ax = plt.subplots()
bottom = np.zeros(len(centroids))

data = {"protein": n_protein,
        "water": n_waters,
        "membrane": n_membrane,
        "ion": n_ion}

for key, vals in data.items():
    p = ax.bar(centroids, vals, 0.8, label=key, bottom=bottom)
    bottom += vals

ax.set_title("Number of Atoms in each Centroid")
ax.legend(loc="upper right")
ax.set_xticks(centroids)
plt.show()
```
